```python
from MINI import Mini
```

```python
nums, k = [1, 2, 4, 9, 3, 5, 6, 7, 8, 10], 2
```

```python
def minSlidingWindow(nums, k):
    n = len(nums)
    if n < k:
        return []
    minVals = []
    for i in range(n - k + 1):
        minVals.append(min(nums[i:i+k]))
    return minVals

def MININATOR(nums, k):
    n = len(nums)
    if n < k:
        return []
    mini = Mini(nums[:k])
    minVals = [mini.getMin()]
    for i in range(1, n - k + 1):
        # print(f'============= {i} =============')
        # print("========")
        # print("REMOVING i - 1", nums[i - 1])
        mini.remove(i - 1)
        # print(mini)
        # print("=========")
        # print("ADDING nums[i + k - 1]", nums[i + k - 1])
        mini.add(i + k - 1, nums[i + k - 1])
        # print("=========")
        # print("MINI STATE:", mini)
        # print("TARGET MINI STATE:", f"nums[i:i+k] = {nums[i:i+k]}")
        mm = mini.getMin()
        # print(mm, "MM")
        minVals.append(mm)
    return minVals
```

```python
print(nums, k)
print("ANSWER", minSlidingWindow(nums, k))
```

```python
print(nums, k)
MININATOR(nums, k)
```

```python
import numpy as np
import time

start = time.time()
lastTime = start
print("STARTING")
resX, resY = [], []
interval = 50
repetitions = 500
maximum = 10000
for i in range(interval, maximum, interval):
    for _ in range(repetitions):
        nums = np.random.randint(0, 100, size=i)
        k = np.random.randint(1, i)

        if minSlidingWindow(nums, k) != MININATOR(nums, k):
            break
    ct = time.time()
    elapsed = ct - lastTime
    averageElapsed = elapsed/repetitions
    print('i:', i, ' ', 'seconds elapsed:', elapsed)
    resX.append(i)
    resY.append(averageElapsed)
    lastTime = ct
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(resX, resY)
plt.xlabel('Input Size')
plt.ylabel('Time (seconds)')
plt.title('Runtime vs Input Size')
plt.grid(True)
plt.show()
```
